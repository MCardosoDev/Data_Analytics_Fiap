```python
dataset['mpg'] = dataset['mpg'].str.replace(',', '.').astype('float')
dataset['cylinders'] = dataset['cylinders'].str.replace(',', '.').astype('float')
dataset['displacement'] = dataset['displacement'].str.replace(',', '.').astype('float')
dataset['horsepower'] = dataset['horsepower'].str.replace(',', '.').astype('float')
dataset['weight'] = dataset['weight'].str.replace(',', '.').astype('float')
dataset['acceleration'] = dataset['acceleration'].str.replace(',', '.').astype('float')
dataset['model year'] = dataset['model year'].str.replace(',', '.').astype('float')
dataset['origin'] = dataset['origin'].str.replace(',', '.').astype('float')
dataset
```
