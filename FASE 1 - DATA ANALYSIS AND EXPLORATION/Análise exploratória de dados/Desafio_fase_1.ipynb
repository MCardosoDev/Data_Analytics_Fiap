### 4.a

```python
dataset_pd['quality'].describe()
```

```python
dataset_pd['fixed acidity'].describe()
```

```python
ax = sns.displot(data=dataset_pd, x="fixed acidity", hue="quality", kde=True, multiple="stack")
ax.set_titles("Frequency of fixed acidity vs. quality")
ax.set_xlabels("fixed acidity")
ax.set_ylabels("Frequency")
ax = ax
```

```python
ax = dataset_pd.plot(x='quality', y='fixed acidity', kind='hist', bins=10, figsize=(20,6))
ax.set_ylabel('Frequency')
ax.set_xlabel('Fixed acidity (mg/L)')
ax.set_title('Frequency of quality vs. fixed acidity')
ax = ax
```

```python
ax = dataset_pd.plot(x='quality', y='fixed acidity', kind='kde', figsize=(20,6))
ax.set_xlabel('Fixed acidity (mg/L)')
ax.set_ylabel('Frequency')
ax.set_title('Frequency of fixed acidity vs. quality')
ax = ax
```

```python
ax = sns.displot(data=dataset_pd, x='fixed acidity', kde=True, kind='hist', stat="density")
ax.figure.set_size_inches(20, 6)
ax.set_titles('Frequency of fixed acidity vs. quality', fontsize=20)
ax.set_xlabels('Fixed acidity (mg/L)', fontsize=16)
ax.set_ylabels('Frequency', fontsize=16)
ax = ax
```

```python
ax = sns.histplot(data=dataset_pd, x='fixed acidity', kde=True, stat='density')
ax.figure.set_size_inches(20, 6)
ax.set_title('Frequency of fixed acidity vs. quality', fontsize=18)
ax.set_ylabel('Fixed acidity (mg/L)', fontsize=14)
ax.set_xlabel('Frequency', fontsize=14)
ax = ax
```

```python
ax = sns.jointplot(x='quality', y='fixed acidity', data=dataset_pd, kind='reg')
ax.fig.suptitle('Frequency of fixed acidity vs. quality', fontsize=18, y=1.05)
ax.set_axis_labels("Fixed acidity (mg/L)", "Quality (mg/L)", fontsize=14)
ax = ax
```
